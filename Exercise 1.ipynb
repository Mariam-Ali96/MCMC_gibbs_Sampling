```r
lakes= c('Winnipeg', 'Winnipegosis','Manitoba','SouthernIndian', 'Cedar', 'Island', 'Gods',  'Cross', 'Playgreen')
elevation = c(217,254,248,254,253,227,178,207,217)
area= c(24387,5374,  4624, 2247, 1353,1223, 1151, 755,657)
```

```r
manitoba.lakes <- data.frame(lakes, elevation, area)
#manitoba.lakes
manitoba.lakes[which.max(manitoba.lakes$elevation),]
manitoba.lakes[which.min(manitoba.lakes$elevation),]
```

```r
Ordered = manitoba.lakes[order(manitoba.lakes$area),]
Ordered
Area_ofLakes = sum(manitoba.lakes$area)
paste("Area covered by Lakes", Area_ofLakes, sep = " : ")
#Area_ofLakes
```
