Density function property to be equal to 1 from -inf to inf is equal to 1, so finding c is by itegrating and receprocating the integral since the function is zero everywhere else

```r
#a)
integrand = function(t) {(t-1)*(2-t)}
integral =integrate(integrand, lower = 1, upper = 2)

c = 1/integral$value
print(c)
```

PDF

```r
dx <- 0.001
x <- seq(1,2, by = dx)
plot(x, integrand(x), type = "l", ylab = "Probability density function", main = "PDF")
```

b:CDF

```r
fx <-integrand
#fx
dx <- 0.001
x <- seq(1,2, by = dx)
plot(x, cumsum(fx(x)), type = "l", ylab = "cummulative probability", main = "CDF")
```

```r
#a)
f2 = function(x) {2/x^2}
dx <- 0.001
x <- seq(1,2, by = dx)
plot(x, f2(x), type = "l", ylab = "Probability density function", main = "PDF")
```

cdf

```r
dx <- 0.001
x <- seq(1,2, by = dx)
plot(x, cumsum(f2(x)), type = "l", ylab = "cummulative probability", main = "CDF")
```

```r
xnew2 <- seq(1,1.5, by = dx)
plot(xnew2, f2(xnew2), type = "l", ylab = "Probability density function", main = "PDF")
```
