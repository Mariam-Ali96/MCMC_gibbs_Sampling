```r
x = c(0,1)
y=c(0.3,0.7)
plot(x,y,type="h",xlim=c(-1,2),ylim=c(0,1),lwd=2,col="blue",ylab="p")
points(x,y,pch=16,cex=2,col="dark red")
```

```r
#Binomial size = 3
p=dbinom(0:3,size=3,prob=1/2)
p
n=3
p=1/2
x=0:3
p=dbinom(x,size=n,prob=p)
plot(x,p,type="h",xlim=c(-1,4),ylim=c(0,1),lwd=2,col="blue",ylab="p")
points(x,p,pch=16,cex=2,col="dark red")

#Binomial size = 10
p=dbinom(0:10,size=10,prob=1/2)
p
n=10
p=1/2
x=0:10
p=dbinom(x,size=n,prob=p)
plot(x,p,type="h",xlim=c(-1,11),ylim=c(0,0.5),lwd=2,col="blue",ylab="p")
points(x,p,pch=16,cex=2,col="dark red")
```

```r
#Exponential Distribution
library(ggplot2)
x = 1:10
pex1 = pexp(x, rate = 1)
pex2 = pexp(x, rate = 1/2)
pex3 = pexp(x, rate = 1/3)
pex3 = pexp(x, rate = 1/3)
pex4 = pexp(x, rate = 1/4)
pex5 = pexp(x, rate = 1/5)

plot( x, pex1, type="l", col="red" )
par(new=TRUE)
plot( x, pex2, type="l", col="green" )
par(new=TRUE)
plot( x, pex3, type="l", col="blue" )
par(new=TRUE)
plot( x, pex4, type="l", col="black" )
par(new=TRUE)
plot( x, pex5, type="l", col="red" )


legend(1, 1, legend=c("rate = 1", "rate = 1/2", "rate = 1/3", "rate = 1/4", "rate = 1/5"), col=c("red","green", "blue", "black","red"), lty=1:7, cex=1)
```
